```python
import seaborn as sns

titanic = sns.load_dataset('titanic')
titanic.head()
```

### 수치형 데이터 시각화

#### histogram

```python
sns.histplot(data=titanic, x='age')
```

```python
sns.histplot(data=titanic, x='age', bins=10)
```

```python
sns.histplot(data=titanic, x='age', bins=10, hue='survived')
```

```python
sns.histplot(data=titanic, x='age', bins=10, hue='survived', multiple='stack')
```

#### kdeplot(커널밀도추정 함수 그래프)
- 히스토그램을 매끄럽게 곡선으로 연결한 그래프

```python
sns.kdeplot(data=titanic, x='age')
```

```python
sns.kdeplot(data=titanic, x='age', hue='alive', multiple='stack')
```

#### displot(분포도)
- 수치형 데이터 하나의 분포를 나타내는 그래프

```python
sns.displot(data=titanic, x='age')
```

```python
sns.displot(data=titanic, x='age', kind='kde')
```

```python
sns.displot(data=titanic, x='age', kde=True)
```

#### rugplot(러그플롯)
- marginal distribution(주변 분포)를 나타내는 그래프

```python
sns.kdeplot(data=titanic, x='age')
sns.rugplot(data=titanic, x='age')
```

### 범주형 데이터 시각화

#### barplot(막대그래프)
- 범주형 데이터 값에 따라 수치형 데이터 값이 어떻게 달라지는지 파악

```python
sns.barplot(x='class', y='fare', data=titanic)
```

#### pointplot(포인트플롯)
- 한 화면에 여러 그래프를 그릴 때 사용하기 좋음

```python
sns.pointplot(x='class', y='fare', data=titanic)
```

#### Boxplot(박스플롯)

```python
sns.boxplot(x='class', y='age', data=titanic)
```

#### violinplot(바이올린플롯)
- 박스플롯과 커널밀도추정 함수 그래프를 합쳐 놓은 그래프

```python
sns.violinplot(x='class', y='age', data=titanic)
```

```python
sns.violinplot(x='class', y='age', hue='sex', data=titanic, split=True)
```

#### countplot(카운트플롯)
- 범주형 데이터의 개수를 확인할 때 사용하는 그래프

```python
sns.countplot(x='class', data=titanic)
```

```python
sns.countplot(y='class', data=titanic)
```

#### pie(파이그래프)
- 범주형 데이터별 비율을 알아볼 때 사용하기 좋은 그래프

```python
import matplotlib.pyplot as plt

x=[10,60,30] # 범주형 데이터별 파이 그래프의 부채꼴 크기(비율)
labels=['A', 'B', 'C']

plt.pie(x=x, labels=labels, autopct='%.1f%%')
```

### 데이터 관계 시각화

#### heatmap(히트맵)
- 데이터간 관계를 색상으로 표현한 그래프로 비교해야 할 데이터가 많을 때 주로 사용

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
import pandas as pd

columns = ["year", "month", "passengers"]
df = pd.DataFrame(flights, columns=columns)
df
```

```python
# flights_pivot = flights.pivot(index='month', columns='year', values='passengers')

flights_pivot = df.pivot('month', 'year', 'passengers')
flights_pivot
```

```python
sns.heatmap(flights_pivot)
```

#### lineplot(라인플롯)
- 두 수치형 데이터 사이의 관계를 나타낼 때 사용

```python
sns.lineplot(x='year', y='passengers', data=flights)
```

#### scatterplot(산점도)

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='time', data=tips)
```

#### regplot(회귀선을 포함한 산점도 그래프)

```python
sns.regplot(x='total_bill', y='tip', data=tips)
```

```python
# 신뢰구간 99%
sns.regplot(x='total_bill', y='tip', ci=99, data=tips) 
```
